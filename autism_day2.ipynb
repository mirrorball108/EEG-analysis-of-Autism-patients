```python
import pandas as pd
import numpy as np
```

```python
df_pre=pd.read_csv('/content/pre_treatement_combined.csv')
```

```python
df_pre.head()
```

```python
df_post=pd.read_csv('/content/post_treatment_combined.csv')
df_post.head()
```

```python
df_post=df_post.drop([0,1])
```

```python
df_post.head()
```

```python
df_pre.size
```

```python
df_pre.shape
```

```python
df_post.shape
```

```python
#creating labels
pre=[1]*7680
post=[0]*7679

print(pre[0:5])
print(post[0:5])

labels=pre+post
print(len(labels))
type(labels)
```

```python
df = pd.concat([df_pre, df_post], ignore_index=True)
```

```python
df.shape
```

```python
#check empty values
df.isna().any(axis=1).sum()
```

```python
#feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_scaled = sc.fit_transform(df)
```

```python
#compute covariance matrix
cov_matrix=np.cov(df_scaled.T)
cov_matrix
```

```python
#calculate eigen values and eigen vectors
eigen_values,eigen_vectors=np.linalg.eigh(cov_matrix)
```

```python
#PCA

from sklearn.decomposition import PCA
pca=PCA(n_components=5)
train_pca=pca.fit_transform(df_scaled)
```

```python
#explain variance ratio

pca.explained_variance_ratio_
```

```python
pca_df=pd.DataFrame(data=train_pca,columns=['PC1','PC2','PC3','PC4','PC5'])
pca_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(pca_df, labels, test_size=0.3, random_state=42)

# Create and train the SVM model
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy}")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
