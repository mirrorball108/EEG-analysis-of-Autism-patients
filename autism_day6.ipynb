## **Poincare Plot**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def poincare_plot(name, signal, type):
    x = signal[:-1]
    y = signal[1:]
    plt.figure(figsize=(6, 6))
    plt.scatter(x, y, alpha=0.6)
    plt.xlabel('x(t)')
    plt.ylabel('x(t+1)')
    plt.title('Poincaré Plot of '+type+ name)
    plt.grid(True)
    plt.axis('equal')
    plt.show()

    x = np.array(signal[:-1])
    y = np.array(signal[1:])
    diff = y - x

    sd1 = np.sqrt(np.var(diff) / 2)
    sd2 = np.sqrt(2 * np.var(signal) - sd1**2)
    print(f"SD1 (Short-term): {sd1:.4f}")
    print(f"SD2 (Long-term): {sd2:.4f}")
```

```python
for i in df_pre.columns:
  poincare_plot(i,df_pre[i], 'pre-treatment ')
  poincare_plot(i,df_post[i], 'post-treatment ')
```
